# Fix Missing Heroicons Dependency

The error indicates that `@heroicons/react` is missing. The import path `@heroicons/react/24/outline` requires Heroicons v2.

Run the following command to install the package:

```python
npm install @heroicons/react
```

After the installation is complete, restart your development server:

```python
npm run dev
```

# Fix: No Space Left on Device (ENOSPC)

The error `ENOSPC` means your disk is full. You need to free up space. 
Run the commands below to clear the npm cache and delete the current `node_modules` folder to reclaim space.

```python
npm cache clean --force
Remove-Item -Recurse -Force node_modules
Remove-Item -Force package-lock.json
```

After clearing space, reinstall dependencies and start the server:

```python
npm install
npm run dev
```
